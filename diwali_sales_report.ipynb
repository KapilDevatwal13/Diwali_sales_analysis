```python
df['Amount']=df['Amount'].astype('int')
```

Analysis

```python
ax=sns.countplot(x='Gender' , data=df , hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
a=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.bar(a['Gender'],a['Amount'], color=['Blue','orange'])
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.grid()
plt.show()
```

---> From above graphs we can says that most of the buyer are Females and even purchasing power of Females are more then Males.

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
ax=sns.countplot(x='Age Group' , data=df ,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

---> By help of above graph we can says that most of the buyers are age_group 26-35  female.

```python
ax=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders' ,ascending=False).head(10)
plt.figure(figsize=(10,5))
a=sns.barplot(data=ax,x='State',y='Orders')
for bars in a.containers:
    a.bar_label(bars)
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.title('Top States With highest Orders')
plt.show()
```

```python
ax=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(12,5))
sns.barplot(data=ax,x='State',y='Amount')
plt.xticks(rotation=45,ha='right')
plt.show()
```

---> From above graph we can see that most of the orders and sales are from Utter Pradesh, Maharashtra, Karnataka respectively.

```python
ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=ax,x='Marital_Status',y='Amount',hue='Gender')
```

--->Above graph shows that most of the buyer are married (Females) and they have high purchasing power.

```python
plt.figure(figsize=(15,5))
plt.xticks(rotation=45)
sns.countplot(x='Occupation',data=df)
```

```python
ax=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
#sns.set(rc={'figure.figsize':(10,5)})
plt.figure(figsize=(20,5))
a=sns.barplot(data=ax,x='Occupation',y='Amount')
```

--->Above graph wants to tell that most of the buyer are working in IT, Healthcare and Aviation.

```python
plt.figure(figsize=(12,4))
plt.xticks(rotation=45,ha='right')

sns.countplot(data=df,x='Product_Category')
```

```python
ax=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(12,4))
plt.xticks(rotation=45,ha='right')
sns.barplot(data=ax,x='Product_Category',y='Amount')
```

--->From the above graphs we can see that sold product are from Food, Clothing and Electronics Category.

```python
ax=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(12,5))
sns.barplot(data=ax,x='Product_ID',y='Orders')
```

```python
ax=df.groupby(['User_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(12,5))
sns.barplot(data=ax,x='User_ID',y='Orders')
ax
```
